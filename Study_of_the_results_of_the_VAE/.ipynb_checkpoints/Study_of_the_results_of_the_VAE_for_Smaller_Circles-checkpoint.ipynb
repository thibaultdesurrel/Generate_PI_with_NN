# Study of the results of the VAE for Smaller Circles

In this notebook we study the results of the VAE trained on the smaller circles dataset.
This are the things we do :
   1. We look at the output of the VAE given testing data,
   2. We try to predict one modality (PI or density) given only the other one,
   3. We study the latent distribution
   4. We draw a random $z_{joint}$ and passe it throw the decoder to see the output
   5. We try to go from a small to a big circle

```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from time import time

import gudhi as gd
import gudhi.representations

from tqdm import tqdm

from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
from scipy.stats import multivariate_normal

from sklearn.neighbors import KernelDensity
```

### We strat by loading the VAE

```python
vae = tf.keras.models.load_model('../Trained_VAE/VAE_full_Smaller_Circles/')
vae.summary()
```

We then load the data and normalize the inputs

```python
data = np.load('../data/PI_data_smaller_circle_density.npz')

PC_train = data["data_train"]
PI_train = data["PI_train"]
density_train = data["density_train"]
PC_test = data["data_test"]
PI_test = data["PI_test"]
density_test = data["density_test"]

N_sets_train = PC_train.shape[0]
N_sets_test = PC_test.shape[0]
N_points = PC_train.shape[1]
PI_size = int(np.sqrt(PI_train.shape[1]))
size = int(np.sqrt(density_test.shape[1]))
dim = 2
batch_size = 64

print("N_sets_train : ", N_sets_train)
print("N_sets_test : ", N_sets_test)
print("N_points : ", N_points)
print("PI_size : ", PI_size)
print("Size : ", size)

PI_train /= max(np.max(PI_train), np.max(PI_test))
PI_test /= max(np.max(PI_test), np.max(PI_test))

for i in range(N_sets_train):
    density_train[i] /= np.max(density_train[i])
for i in range(N_sets_test):
    density_test[i] /= np.max(density_test[i])
```

### Try to predict a couple of (PI,density) given only one of the two modalities

#### Predict a density from a PI

To do this we give the PIs encoder the PIs

```python
z_mean_PI, z_var_PI = vae.encoder_PI(PI_test)
```

From the $(\mu_{PI},\sigma_{PI})$ we get from the PIs encoder we can compute $(\mu_{joint},\sigma_{joint})$

```python
mean_joint = (z_mean_PI / z_var_PI) * (1 + 1 / z_var_PI)
var_joint = (1 + 1 / z_var_PI)
```

We then sample $z_{joint}$ from $\mathcal N(\mu_{joint},\sigma_{joint})$

```python
z_joint = vae.sampling((mean_joint, var_joint))
```

We can then decode $z_{joint}$ to a couple of (PI,density)

```python
z_PI, z_PC = tf.split(vae.shared_decoder(z_joint),
                      num_or_size_splits=2,
                      axis=1)

reconstructed_PI = vae.decoder_PI(z_PI)
reconstructed_PC = vae.decoder_PC(z_PC)
```

We can plot them

```python
for i in range(4):
    plt.subplot(4, 2, 2 * i + 2)
    plt.imshow(np.flip(
        np.reshape(reconstructed_PI[i + 10], [PI_size, PI_size]), 0),
               vmin=0,
               vmax=1,
               cmap='jet')
    plt.colorbar()
    plt.axis('off')
    plt.subplot(4, 2, 2 * i + 1)
    plt.imshow(np.flip(np.reshape(reconstructed_PC[i + 10], [size, size]), 0),
               cmap='jet')
    #plt.colorbar()
    plt.axis('off')
#plt.savefig("Rsultats/VAE_full_2_predict_density_from_PI.pdf")
```

#### Predict a PI from a density 

Here we do the same thing but with a density at the start : we give the PCs encoder the densities

```python
z_mean_PC, z_var_PC = vae.encoder_PC(density_test)
```

From the $(\mu_{PC},\sigma_{PC})$ we get from the PCs encoder we can compute $(\mu_{joint},\sigma_{joint})$

```python
mean_joint = (z_mean_PC / z_var_PC) * (1 + 1 / z_var_PC)
var_joint = (1 + 1 / z_var_PC)
```

We then sample $z_{joint}$ from $\mathcal N(\mu_{joint},\sigma_{joint})$

```python
z_joint = vae.sampling((mean_joint, var_joint))
```

We can then decode $z_{joint}$ to a couple of (PI,density)

```python
z_PI, z_PC = tf.split(vae.shared_decoder(z_joint),
                      num_or_size_splits=2,
                      axis=1)

reconstructed_PI = vae.decoder_PI(z_PI)
reconstructed_PC = vae.decoder_PC(z_PC)
```

We can plot them

```python
for i in range(4):
    plt.subplot(4, 2, 2 * i + 2)
    plt.imshow(np.flip(
        np.reshape(reconstructed_PI[i + 10], [PI_size, PI_size]), 0),
               vmin=0,
               vmax=1,
               cmap='jet')
    plt.colorbar()
    plt.axis('off')
    plt.subplot(4, 2, 2 * i + 1)
    plt.imshow(np.flip(np.reshape(reconstructed_PC[i + 10], [size, size]), 0),
               cmap='jet')
    #plt.colorbar()
    plt.axis('off')
#plt.savefig("Rsultats/VAE_full_2_predict_PI_from_density.pdf")
```

### Study of the latent distribution 

To study the latent distribution we compute the $(\mu_{joint},\sigma_{joint})$ for data representing circle of different size.

We can then encode these gaussians into a `multivariate_normal` object and then plot the gaussian in 3D or in 2D.

```python
def create_circle(N_points, r, x_0, y_0):
    X = []
    for i in range(N_points):  # On fait un cercle
        theta = np.random.uniform() * 2 * np.pi
        X.append([(r * np.cos(theta)) + x_0, (r * np.sin(theta) + y_0)])
    return np.array(X)
```

```python
size = 50
PI_size = 50
x_sample = np.linspace(-1, 1, size)
y_sample = np.linspace(-1, 1, size)
#xx, yy = np.meshgrid(x_sample, y_sample)
coord = np.array([[x, y] for x in x_sample for y in y_sample])
```

```python
N = 100
nb_cercles = 2
N_points = 500

PI_rep = gd.representations.PersistenceImage(bandwidth=1e-1,
                                             weight=lambda x: x[1]**(1),
                                             resolution=[PI_size, PI_size],
                                             im_range=[0, 1, 0, 1])

r_liste = np.linspace(0.1, 1, N)
mu_liste = []
sigma_liste = []

for j in tqdm(range(N)):
    #print("i = ", j, "r = ", r_liste[j])
    circles = np.zeros((nb_cercles, N_points, 2))
    density = np.zeros((nb_cercles, size * size))
    PI = np.zeros((nb_cercles, PI_size * PI_size))
    for i in (range(nb_cercles)):
        circles[i] = create_circle(N_points, r_liste[j], 0, 0)

        kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(circles[i])
        density[i] = np.exp(kde.score_samples(coord))

        rcX = gd.AlphaComplex(points=circles[i]).create_simplex_tree()
        dgmX = rcX.persistence()
        pi = PI_rep.fit_transform([rcX.persistence_intervals_in_dimension(1)])
        PI[i] = pi[0]

    z_mean_PC, z_var_PC = vae.encoder_PC(density)
    z_mean_PI, z_var_PI = vae.encoder_PI(PI)

    mean_joint = (z_mean_PI / z_var_PI +
                  z_mean_PC / z_var_PC) * (1 + 1 / z_var_PI + 1 / z_var_PC)
    var_joint = (1 + 1 / z_var_PI + 1 / z_var_PC)

    #print("mean_joint = ",mean_joint)
    #print("var_joint = ",var_joint)

    mu_liste.append(np.mean(mean_joint, 0))
    sigma_liste.append(np.mean(var_joint, 0))
```

```python
M = 100
X, Y = np.meshgrid(np.linspace(0, 35, M), np.linspace(-20, 8, M))
d = np.dstack([X, Y])

gauss = []
Z = []

for i in range(len(mu_liste)):
    gauss.append(multivariate_normal(mean=mu_liste[i], cov=sigma_liste[i]))

    Z.append(gauss[i].pdf(d).reshape(M, M))
```

Plot in 3D

```python
fig = plt.figure(figsize=(12, 12))
ax = Axes3D(fig)

for i in range(len(mu_liste)):
    ax.plot_surface(X, Y, Z[-i], rstride=1, cstride=1, cmap='RdPu')
#carre = ax.plot_surface(X,
#                        Y,
#                        Z_1,
#                        rstride=1,
#                        cstride=1,
#                        alpha=0.8,
#                        cmap='seismic')
#cercle = ax.plot_surface(X,
#                         Y,
#                         Z_2,
#                         rstride=1,
#                         cstride=1,
#                         alpha=0.8,
#                         cmap='jet')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#ax.set_xlim(-5, 5)
#ax.set_ylim(-5, 5)

#cbar1 = plt.colorbar(carre, location='bottom', shrink=0.5, pad=0.001)
#cbar2 = plt.colorbar(cercle,
#                     label='Latent distribution of the circle',
#                     location='bottom',
#                     shrink=0.5,
#                     pad=0.001)
#cbar1.set_label('Latent distribution of the square', size=18)
#cbar2.set_label('Latent distribution of the circle', size=18)

plt.title('3D representation of the latent distribution', size=18)
plt.show()

#fig.savefig("Rsultats/VAE_full_2_distributions_3D.pdf")
```

### Generate random $z_{joint}$

We try to generate a random $z_{joint}$ from $\mathcal N((15,-4),\begin{pmatrix}15 &0 \\ 0 & 4 \end{pmatrix})$. We then decode this random $z_{joint}$ and plot the PI and density we get from it.

```python
z_joint = np.random.normal(size=(4, 2),loc=[15,-4],scale=[15,4])
z_PI, z_PC = tf.split(vae.shared_decoder(z_joint),
                      num_or_size_splits=2,
                      axis=1)

reconstructed_PI = vae.decoder_PI(z_PI)
reconstructed_PC = vae.decoder_PC(z_PC)
```

```python
for i in range(4):
    plt.subplot(4, 2, 2 * i + 2)
    plt.imshow(np.flip(np.reshape(reconstructed_PI[-i], [PI_size, PI_size]),
                       0),
               vmin=0,
               vmax=1,
               cmap='jet')
    plt.colorbar()
    plt.axis('off')
    plt.subplot(4, 2, 2 * i + 1)
    plt.imshow(np.flip(np.reshape(reconstructed_PC[-i], [size, size]), 0),
               cmap='jet')
    #plt.colorbar()
    plt.axis('off')
```

### Go from a square to a circle 

To go from a smal to a big circle, we arrange N $z_{joint}$ from $(0,0)$ (representing a small circle) to $(0,20)$ representing a big circle. We can then decode those $z_{joint}$ and plot the output.

```python
PI_size = 50
size = 50
N = 20
```

```python
z_joint = np.stack((np.transpose(np.linspace(0, 20, N)), np.zeros((N, ))), 1)
```

```python
z_PI, z_PC = tf.split(vae.shared_decoder(z_joint),
                      num_or_size_splits=2,
                      axis=1)

reconstructed_PI = vae.decoder_PI(z_PI)
reconstructed_PC = vae.decoder_PC(z_PC)
```

```python
for i in range(N):
    plt.figure(figsize=(4, 8))
    plt.subplot(2, 1, 1)
    plt.imshow(np.flip(np.reshape(reconstructed_PI[i], [PI_size, PI_size]), 0),
               vmin=0,
               vmax=1,
               cmap='jet',
               aspect='equal')
    #plt.colorbar()
    plt.axis('off')
    plt.subplot(2, 1, 2)
    plt.imshow(np.flip(np.reshape(reconstructed_PC[i], [size, size]), 0),
               cmap='jet',
               aspect='equal')
    #plt.colorbar()
    plt.axis('off')
    #plt.savefig("Rsultats/from_square_to_circle_"+str(i)+".png")
```
